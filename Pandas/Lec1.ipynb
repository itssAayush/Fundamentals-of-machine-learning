Series in Pandas

SERIES FROM DICTIONARY

```python
data = {
    "a": 10,
    "b": 20,
    "c": 30
}

s = pd.Series(data)

print(s)
```

DataFrames in Pandas

```python
data = {
    "Name": ["Aman", "krish", "karan"],
    "Age": [22, 23, 21],
    "Marks": [85, 90, 88]
}
```

```python
print(data)
```

```python
type(data)
```

```python
marks = pd.DataFrame(data)
```

```python
print(marks)
```

```python
marks = pd.DataFrame(
    data, 
    index =["Student1", "Student2", "Student3"],
)
```

```python
print(marks)
```

```python
marks.set_index("Name", inplace=True) # if we are setting iplace = true it means we are changing the original object itself.
```

```python
print(marks)
```
