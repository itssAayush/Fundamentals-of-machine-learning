Slicing & Indexing

```python
import numpy as np

sales = np.array([
    [50, 65, 70, 60],   
    [80, 75, 90, 85],   
    [40, 55, 60, 65],   
    [30, 45, 50, 55]    
])

print(sales)

print(sales[0]) #print first row

print(sales[:, 0:2])

print(sales[2][3])

print(sales[0:2, 0:2])
```

Reshaping & Linespace

```python
sales = np.array([
    [120, 150, 170, 160],
    [200, 210, 190, 220],
    [90,  100, 110, 105]
])

flatten = sales.reshape(12)
print(flatten) #flatten

reshape = sales.reshape(2, 6)
print(reshape)
```
