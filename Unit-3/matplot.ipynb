DATA VISUALISATION

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    "Name" : ["Aayush", "Shubham", "Srishti", "Hardin", "Rohan", "Karan", "Neha"],
    "Age" : [21, 22, 34, 12, 45, 23, 42],
    "Gender" : ["Male", "Male", "Female", "Male", "Male", "Male", "Female"],
    "Marks" : [99, 23, 91, 20, 56, 75, 89]
}

df = pd.DataFrame(data)
```

Line Plot for Marks Trend

```python
plt.plot(df["Marks"])

plt.title("Marks Trend")
plt.xlabel("Students")
plt.ylabel("Marks")

plt.show()
```

Bar Plot for comparison

```python
plt.bar(df["Name"], df["Marks"])

plt.title("Marks of students")
plt.xlabel("Student name")
plt.ylabel("Marks")

plt.show()
```

Histogram for distribution of age


```python
plt.hist(df["Age"], bins = 5)

plt.title("Age distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")

plt.show()
```

Scatter plot

Relationship between age and marks


```python
plt.scatter(df["Age"], df["Marks"])

plt.title("Age vs Marks")
plt.xlabel("Age")
plt.ylabel("Marks")

plt.show()
```

Pie Chart
Gender Distribution

```python
gender_counts = df["Gender"].value_counts()
plt.pie(gender_counts, labels=gender_counts.index, autopct="%1.1f%%")
plt.title("Gender Distribution")
plt.show()
```

Customizing plots

Example with colors and grid

```python
plt.bar(df["Name"], df["Marks"], color="cyan")

plt.title("Student Marks")
plt.xlabel("Student Name")
plt.ylabel("Marks")

plt.grid(True)
plt.show()
```

Show Two Graphs Together

```python
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.hist(df["Age"])
plt.title("Age Distribution")

plt.subplot(1,2,2)
plt.bar(df["Name"], df["Marks"])
plt.title("Marks of students")

plt.show()
```
