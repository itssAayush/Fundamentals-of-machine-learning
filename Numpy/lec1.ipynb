basic array operations (vector power)

```python
a = np.array([1,2,3])
b = np.array([4,5,6])
```

```python
print(a+b)
print(a-b)
print(a*b)
print(a/b)
```

```python
mat = np.array([[1,2,3],[4,5,6]])
print(mat)
```

```python
print(a + 10)
print(a * 2)
```

Universal Functions

```python
arr = np.array([1,4,9,16])

print(np.sqrt(arr))
print(np.log(arr))
print(np.exp(arr))
print(np.max(arr))
print(np.min(arr))
print(np.mean(arr))
print(np.sum(arr))
print(np.std(arr))
```
